#  다중 클래스 분류기
- 3개 이상의 클래스(범주)가 있는 경우

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
```

```python
### data load
iris = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/datasets/iris.csv")
iris
```

```python
iris.loc[50:,]
```

```python
### 설명변수(x)와 반응변수(y)
x = iris[["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]]
y = iris["Species"]
```

```python
### Train & Test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4)
```

```python
### logistic regression
# multi_class = {'auto','ovr','multinomial'}, default= 'auto'
logistic = LogisticRegression(multi_class = 'ovr')
model = logistic.fit(x_train,y_train)
```

```python
### 예측
y_pred = model.predict(x_test)
y_pred
```

```python
### 교차표
pd.crosstab(y_test, y_pred) # 행이름 : y_test
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

###  house 가격 예측하기

```python
# 설명변수 후보  bedrooms, bathrooms, sqft_living, view, condition,sqft_above, yr_built, yr_renovated
# 반응변수 price_cat
```

```python
house = pd.read_csv('http://youngho.iwinv.net/data/house_price_prediction.csv')
house.head()
```

```python
house.describe()
```

```python
### 설명변수(x)와 반응변수(y)
x = house[['bedrooms', 'bathrooms','yr_renovated', 'view', 'yr_built']]
#설명변수가 많아지면 accuracy가 증가한다.
# accuracy를 증가시키려면  가격에 영향을 많이 주는 변수를 잘 찾아양 한다.
y = house["price_cat"]
```

```python
### Train & Test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4)
```

```python
### logistic regression
# multi_class = {'auto','ovr','multinomial'}, default= 'auto'
logistic = LogisticRegression(multi_class = 'ovr')
model = logistic.fit(x_train,y_train)
```

```python
### 예측
y_pred = model.predict(x_test)
y_pred
```

```python
### 교차표
pd.crosstab(y_test,y_pred)
```

```python
### accuracy
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# accuracy
model.score(x_test,y_test)
```
