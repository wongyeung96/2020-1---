## 단순 선형 회귀분석

```python
### library
import pandas as pd
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
# 자료 입력
x = [168,160,170,158,176,161,180,183,180,167,179,171,166]  # 아버지의 키
y = [179,169,180,160,178,170,183,187,179,172,181,173,165]  # 아들의 키
df = pd.DataFrame({"x":x,"y":y})
df
```

```python
# 변수별 자료 요약
df.describe()
```

```python
# 산점도 행렬
p = pd.plotting.scatter_matrix(df, figsize = (8,8), marker = 'o')
```

```python
### 단순 선형 회귀분석
# Regression : y ~ x
reg_simp = ols('y ~ x', data =df).fit()
reg_simp.summary()
```

```python
# 산점도와 회귀직선
plt.scatter(df['x'],df['y'])
plt.plot(df['x'],reg_simp.predict(),'red')
plt.show()
```

## 다중 선형 회귀분석 

▣문제Icecream 자료는 1951–03–18에서 1953–07–11까지 4주단위로아이스크림소비를조사한자료로 1인당아이스크림소비금액(cons), 주당평균가족소득(income), 아이스크림판매가격(price), 평균기온(temp)이포함되어 있다.

```python
### CSV 파일에있는자료읽기
import pandas as pd
data_file="https://vincentarelbundock.github.io/Rdatasets/csv/Ecdat/Icecream.csv"
Icecream=pd.read_csv(data_file)
Icecream=Icecream.drop(Icecream.columns[0], axis=1)
Icecream
```

```python
# 변수별 자료 요약
Icecream.describe()
```

```python
# 산점도 행렬
p=pd.plotting.scatter_matrix(Icecream, figsize=(8, 8), marker='o')
```

##### 단순 선형 회귀분석

- 아이스크림 소비 금애과 기온과의 회귀모형

```python
### 단순선형회귀분석
# Regression: 소비금액(cons) ~ 평균기온(temp)
reg_1=ols('cons ~ temp', data=Icecream).fit()
reg_1.summary()
```

```python
# 산점도와회귀직선
plt.scatter(Icecream['temp'],Icecream['cons'])
plt.plot(Icecream['temp'],reg_1.predict(), 'red')
plt.show()
```

```python
### 단순선형회귀분석
# Regression 2: 소비금액(cons) ~ 주당평균가족소득(income)
reg_2=ols('cons ~ income', data=Icecream).fit()
reg_2.summary()
```

```python
### 다중선형회귀분석
# Regression 3: cons ~ temp + income
reg_3=ols('cons ~ temp + income', data=Icecream).fit()
reg_3.summary()
```

```python
### 다중선형회귀분석
# Regression 4: cons ~ temp + income + price
reg_4=ols('cons ~ temp + income + price', data=Icecream).fit()
reg_4.summary()
```
